### Accuracy Comparison: All Features vs. no_ner

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['NB', 'LG', 'DT', 'RF', 'SVM', 'XGB', 'GB', 'KNN']
features = ['all_features', 'no_ner']

# Accuracy values for each feature and model
accuracy_values = {
    'NB': [0.894625, 0.8946807],
    'LG': [0.89675, 0.89675],
    'DT': [0.772125, 0.77],
    'RF': [0.86, 0.8595],
    'SVM': [0.8955, 0.89525],
    'XGB': [0.879875, 0.879375],
    'GB': [0.889375, 0.890375],
    'KNN': [0.762625, 0.762875]
}

# Convert data to arrays
all_features_acc = np.array([accuracy_values[model][0] for model in models])
no_ner_acc = np.array([accuracy_values[model][1] for model in models])

# Set the width of the bars
bar_width = 0.35

# Set the position of the bars on the x-axis
r1 = np.arange(len(models))
r2 = [x + bar_width for x in r1]

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(r1, all_features_acc, color='b', width=bar_width, edgecolor='grey', label='all_features')
plt.bar(r2, no_ner_acc, color='r', width=bar_width, edgecolor='grey', label='no_ner')

# Add labels and title
plt.xlabel('Models', fontweight='bold')
plt.ylabel('Accuracy', fontweight='bold')
plt.xticks([r + bar_width / 2 for r in range(len(models))], models)
plt.title('Accuracy Comparison: All Features vs. no_ner', fontweight='bold')

# Add legend
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
