```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
count_dict = {
    'NB': {
        'accuracy': 0.90638,
        'precision': 0.90633,
        'recall': 0.90638,
        'f1': 0.90625,
        },
    'DTC': {
        'accuracy': 0.90313,
        'precision': 0.90309,
        'recall':  0.90313,
        'f1': 0.90288,
        },
    'LR': {
        'accuracy': 0.90388,
        'precision': 0.90369,
        'recall': 0.90388,
        'f1': 0.90366,
        },
    'RF': {
        'accuracy': 0.87175,
        'precision': 0.87173,
        'recall': 0.87175,
        'f1': 0.87134,
        },
    'XGB': {
        'accuracy': 0.89513,
        'precision': 0.89497,
        'recall': 0.89513,
        'f1': 0.89497,
        }
}

tfidf_dict = {
    'NB': {
        'accuracy': 0.90313,
        'precision': 0.90309,
        'recall': 0.90313,
        'f1': 0.90288,
        },
    'DTC': {
        'accuracy': 0.77688,
        'precision': 0.77609,
        'recall':  0.77688,
        'f1': 0.77642,
        },
    'LR': {
        'accuracy': 0.89188,
        'precision': 0.89171,
        'recall': 0.89188,
        'f1': 0.89118,
        },
    'RF': {
        'accuracy': 0.87525,
        'precision': 0.87492,
        'recall': 0.87525,
        'f1': 0.87473,
        },
    'XGB': {
        'accuracy': 0.89563,
        'precision': 0.89539,
        'recall': 0.89563,
        'f1': 0.895407,
        }
}
```

```python
def plot_model_comparison(colors, metric):
    # Set the y-axis limit to show only the top half
    plt.ylim(0.7, 1.0)

    # Get the model names
    model_names = list(count_dict.keys())

    # Get the accuracies from the dictionaries
    count_accuracies = [count_dict[model][metric] for model in model_names]
    tfidf_accuracies = [tfidf_dict[model][metric] for model in model_names]

    # Plot the bar chart
    plt.bar(np.arange(len(model_names)), count_accuracies, color=colors[0], label='Count')
    plt.bar(np.arange(len(model_names)), tfidf_accuracies, color=colors[1], label='TF-IDF')

    # Add labels and title
    plt.xlabel('Model')
    plt.ylabel(f'{metric.capitalize()}')
    plt.title(f'Comparison of Model {metric.capitalize()} with Count and TF-IDF Vectorization')

    # Set the x-axis tick labels to the model names
    plt.xticks(np.arange(len(model_names)), model_names)

    plt.legend()

    # Show the plot
    plt.show()
```

accuracy

```python
plot_model_comparison(['lightblue', 'darkblue'], 'accuracy')
```

precision

```python
plot_model_comparison(['lightcoral', 'darkred'], 'precision')
```

recall

```python
plot_model_comparison(['lightgreen', 'darkgreen'], 'recall')
```
